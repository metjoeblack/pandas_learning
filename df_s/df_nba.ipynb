## Selecting Rows

```python
nba.loc["LeBron James"]
```

```python
nba.loc[["Kawhi Leonard", "Paul George"]]
```

```python
# We can use loc to extract a sequence of index labels. The syntax mirrors Python’s list slicing syntax. 
# We provide the starting value, a colon, and the ending value. For extractions like this one, 
# I strongly recommended sorting the index first, as it accelerates the speed with which pandas finds the value.

nba.sort_index(ascending=True).loc["Otto Porter":"Patrick Beverley"]
```

```python
nba.iloc[300]
```

```python
nba.iloc[[100, 200, 300, 400]]
```
