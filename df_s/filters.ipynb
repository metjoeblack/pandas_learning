## Caution

Updating a DataFrame column works similarly to setting a key-value pair in a dictionary. If a column with the specified name exists, pandas overwrites it with the new Series. If the column with the name does not exist, pandas creates a new Series and appends it to the right of the DataFrame. The library matches rows in the Series and DataFrame by shared index labels.

```python
# we create a new Series, then we overwrite our existing Mgmt column

employees["Mgmt"] = employees["Mgmt"].astype(bool)
```

```python
employees
```

```python
employees.info()
```

```python
try:
    employees["Salary"] = employees["Salary"].astype(int)
except pd.errors.IntCastingNaNError:
    employees["Salary"] = employees["Salary"].fillna(0).astype(int)
```

### Invesion with ~

```python
my_boolean_series = pd.Series([True, False, True])
```

```python
my_boolean_series
```

```python
~my_boolean_series
```

```python
earning_less_100k = employees["Salary"] < 100_000

employees[earning_less_100k].head()
```

```python
earning_greater_equal_100k = employees[~earning_less_100k]
```

```python
earning_greater_equal_100k.head()
```

```python
all_star_teams = ["Sales", "Legal", "Marketing"]

on_all_star_teams = employees["Team"].isin(all_star_teams)

employees[on_all_star_teams].head(8)
```

```python
between_80k_and_90k = employees["Salary"].between(80_000, 90_000)

employees[between_80k_and_90k].head()
```

```python
eighties_folk = employees["Start Date"].between(
    left="1980-01-01",
    right="1990-01-01",
)
employees[eighties_folk].head()
```

```python
name_starts_with_r = employees["First Name"].between("R", "S")

employees[name_starts_with_r]
```
