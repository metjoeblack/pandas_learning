```python
food_data = {
    "Item": ["Banana", "Cucumber", "Orange", "Tomato", "Watermelon"],
    "Type": ["Fruit", "Vegetable", "Fruit", "Vegetable", "Fruit"],
    "Price": [0.99, 1.25, 0.25, 0.33, 3.00],
}

supermarket = pd.DataFrame(data=food_data)

supermarket
```

```python
groups_obj = supermarket.groupby(by="Type")

groups_obj
```
