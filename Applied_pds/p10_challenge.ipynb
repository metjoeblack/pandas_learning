```python
import pandas as pd
import numpy as np
```

```python
week1 = pd.read_csv("restaurant/week_1_sales.csv")

week1.head()
```

```python
week2 = pd.read_csv("restaurant/week_2_sales.csv")

week2.head()
```

```python
all_weeks = pd.concat(objs=[week1, week2], keys=["Week 1", "Week 2"])

all_weeks
```

```python
ate_both_weeks = week1.merge(
    week2,
    how="inner",
    on="Customer ID",
)

ate_both_weeks.drop_duplicates(subset=["Customer ID"]).head(10)
```

```python
ate_both_weeks[ate_both_weeks["Food ID_x"] == ate_both_weeks["Food ID_y"]]
```

```python
both_weeks_same_items = week1.merge(
    week2,
    how="inner",
    on=["Customer ID", "Food ID"],
)

both_weeks_same_items
```

```python
outer_join = week1.merge(
    week2,
    how="outer",
    on="Customer ID",
    indicator=True,
)

outer_join[outer_join["_merge"] == "both"]
```

```python
outer_join[~(outer_join["_merge"] == "both")]
```
