```python
deliveries.head()
```

```python
deliveries.dtypes
```

```python
(deliveries["delivery_date"] - deliveries["order_date"]).head()
```

```python
deliveries["duration"] = deliveries["delivery_date"] - deliveries["order_date"]
```
