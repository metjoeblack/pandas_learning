```python
video_game_sales = pd.read_csv("cha08_datas/video_game_sales.csv")

video_game_sales
```

```python
video_game_sales.melt(id_vars="Name", value_vars="NA", value_name="Sold").head(8)
```

```python
regions = ["NA", "EU", "JP", "Other"]

video_game_sales_by_region = video_game_sales.melt(
    id_vars="Name", value_vars=regions, var_name="Region", value_name="Sales",
)

video_game_sales_by_region
```

```python
video_game_sales_by_region.pivot_table(
    index="Name", values="Sales", aggfunc="sum",
).head()
```
