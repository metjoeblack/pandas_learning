```python
import pandas as pd
import numpy as np
```

```python
cars = pd.read_csv("cha08_datas/used_cars.csv")

cars.head(10)
```

```python
cars.info()
```

```python
cars["Fuel"].unique()
```

```python
cars["Fuel"] = cars["Fuel"].fillna("NaN Value")
```

```python
cars["Fuel"].unique()
```

```python
cars.nunique()
```

```python
cars["Transmission"] = cars["Transmission"].fillna("NaN Value")
```

```python
cars["Transmission"].unique()
```

```python
cars["Fuel"] = cars["Fuel"].astype(dtype="category")
```

```python
cars["Transmission"] = cars["Transmission"].astype(dtype="category")
```

```python
cars["Manufacturer"] = cars["Manufacturer"].astype(dtype="category")
```

```python
cars.info()
```

```python
cars.nunique()
```

```python
cars["Fuel"].dtype
```

```python
cars.pivot_table(
    index="Fuel",
    values=["Price"],
    aggfunc="sum",
    observed=False,
)
```

```python
cars.pivot_table(
    index="Manufacturer",
    values=["Price"],
    columns="Transmission",
    aggfunc="count",
    fill_value=0,
    margins=True,
    margins_name="Total",
    observed=False,
)
```

```python
cars.pivot_table(
    index=["Year", "Fuel"],
    values=["Price"],
    columns=["Transmission"],
    aggfunc="mean",
    fill_value=0,
    observed=False,
)
```

```python
cars.pivot_table(
    index=["Year", "Fuel"],
    values=["Price"],
    columns="Transmission",
    aggfunc="mean",
    fill_value=0,
    observed=False,
).stack(future_stack=True).head(20)
```
